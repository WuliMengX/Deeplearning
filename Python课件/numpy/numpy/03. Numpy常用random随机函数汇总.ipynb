##  Numpy常用random随机函数汇总

官方文档地址：https://docs.scipy.org/doc/numpy-1.14.0/reference/routines.random.html


```python
import numpy as np
np.random.seed(666)
```

### 1. rand(d0, d1, ..., dn)

返回数据在[0, 1)之间

```python
np.random.rand(5)
```

```python
np.random.rand(3, 4)
```

```python
np.random.rand(2, 3, 4)
```

### 2. randn(d0, d1, ..., dn)
返回数据具有标准正态分布（均值0，标准差1）

```python
np.random.randn(5)
```

```python
np.random.randn(3, 4)
```

```python
np.random.randn(2, 3, 4)
```

### 3. randint(low[, high, size, dtype])

生成随机整数，包含low，不包含high  
如果high不指定，则从[0, low)中生成数字

```python
import numpy as np
np.random.randint(3)
```

```python
np.random.randint(1, 10)
```

```python
np.random.randint(10, 30, size=(5,))
```

```python
np.random.randint(10, 30, size=(2,3,4))
```

### 4. random([size])	
生成[0.0, 1.0)的随机数

```python
np.random.random(5)
```

```python
np.random.random(size=(3,4))
```

```python
np.random.random(size=(2,3,4))
```

### 5. choice(a[, size, replace, p])
a是一维数组，从它里面生成随机结果


```python
# 这时候，a是数字，则从range(5)中生成，size为3
np.random.choice(5, 3)
```

```python
np.random.choice(5, (2,3))
```

```python
# 这时候，a是数组，从里面随机取出数字
np.random.choice([2, 3, 6, 7, 9], 3)
```

```python
np.random.choice([2, 3, 6, 7, 9], (2,3))
```

### 6. shuffle(x)	
把一个数组x进行随机排列

```python
a = np.arange(10)
np.random.shuffle(a)
a
```

```python
a = np.arange(20).reshape(4,5)
a
```

```python
# 如果数组是多维的，则只会在第一维度(这里例子里就是行维度)打散数据
np.random.shuffle(a)
a
```

### 7. permutation(x)
把一个数组x进行随机排列，或者数字的全排列

```python
# 这时候，生成range(10)的随机排列
np.random.permutation(10)
```

```python
# 这时候，在第一维度进行打散（第一个维度一般为按行打散）
arr = np.arange(9).reshape((3, 3))
arr
```

```python
# 注意，这里不会更改原来的arr，会返回一个新的copy
np.random.permutation(arr)
```

```python
arr
```

关于shuffle和permutation的区别这里说两点：

1. shuffle没有返回值，直接在原来的列表上进行打乱排序；而permutation会对一个int行的数值返回一个打乱的列表。

2.由于permutation会复制数据，所以当数据量特别大的时候，使用shuffle的效率更高。

### 8. normal([loc, scale, size])
按照平均值loc和标准差scale生成高斯分布的数字

```python
np.random.normal(1, 10, 10)
```

```python
np.random.normal(1, 10, (3,4))
```

### 9. uniform([low, high, size]) 	
在[low, high)之间生成均匀分布的数字，功能：从一个均匀分布[low,high)中随机采样，注意定义域是左闭右开，即包含low，不包含high.

```python
np.random.uniform(1, 10, 10)
```

```python
np.random.uniform(1, 10, (3,4))
```

### 实例：对数组加入随机噪声

```python
import matplotlib.pyplot as plt
```

```python
# 绘制sin曲线
x = np.linspace(-10, 10, 100)#最小数，最大数，生成100个点
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
# 加入噪声
x = np.linspace(-10, 10, 100)
y = np.sin(x) + np.random.rand(len(x))
plt.plot(x, y)
plt.show()
```
